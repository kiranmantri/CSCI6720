```python
import sys
import time
import io
import numpy as np
import pandas as pd
import select

from PIL import Image
import cozmo

import psycopg2 as pg
import psycopg2.extensions as pgext

import matplotlib.pyplot as plt

dbConn = pg.connect(host="localhost", user="user", password="p4ssw0rd", dbname="CSCI6720G")
dbConn.set_isolation_level(pgext.ISOLATION_LEVEL_AUTOCOMMIT)


def program(robot: cozmo.robot.Robot):
    robot.camera.image_stream_enabled = True        

    for a in range(0,5):
        img = robot.world.latest_image
        if img is None:
            time.sleep(0.1)
            continue
        img = img.raw_image
        imgByteArr = io.BytesIO()        
        img.save(imgByteArr, format='PNG')

        print(imgByteArr)
#         cursor = dbConn.cursor()        
#         cursor.execute("""   INSERT INTO "Data" ("Image") VALUES ({})  """.format(imgByteArr))

        
        plt.imshow(img)
        plt.show()
        plt.close()    
```

```python
cozmo.robot.Robot.drive_off_charger_on_connect = False  # Cozmo can stay on his charger for this example
cozmo.run_program(program, use_viewer=False)
```
